```python
# calculo de ultima cifra octubre 25 determinar si es par o impar dorado tarde
import pandas as pd

# Datos proporcionados
data = {
    "Día": list(range(1, 39)),
    "Cifra": [3, 4, 2, 7, 3, 3, 1, 1, 7, 3, 9, 6, 3, 3, 3, 1, 7, 4, 5, 5,
              5, 2, 4, 7, 4, 8, 4, 4, 5, 8, 9, 2, 5, 2, 7,9,5, 6]
}

# Crear DataFrame
df = pd.DataFrame(data)
df['Par/Impar'] = df['Cifra'].apply(lambda x: 'Par' if x % 2 == 0 else 'Impar')

# Contar ocurrencias de Par e Impar
count_par = df[df['Par/Impar'] == 'Par'].shape[0]
count_impar = df[df['Par/Impar'] == 'Impar'].shape[0]

# Probabilidad de que sea Par o Impar
prob_par = count_par / 32
prob_impar = count_impar / 32

# Mostrar resultados
print(f"Frecuencia de días par: {count_par}")
print(f"Frecuencia de días impar: {count_impar}")
print(f"Probabilidad de que el día 33 sea Par: {prob_par:.2%}")
print(f"Probabilidad de que el día 33 sea Impar: {prob_impar:.2%}")

# Secuencia reciente
print("\nÚltimos días:")
print(df.tail(5))
```

```python
# calculo de ultima cifra octubre 25 numeros impares dorado tarde
import pandas as pd

# Datos proporcionados
data = {
    "Día": list(range(1, 39)),
    "Cifra": [3, 4, 2, 7, 3, 3, 1, 1, 7, 3, 9, 6, 3, 3, 3, 1, 7, 4, 5, 5,
              5, 2, 4, 7, 4, 8, 4, 4, 5, 8, 9, 2, 5,2, 7,9,5, 6]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Filtrar los números de interés
numeros_interes = [1, 3, 5, 7, 9]
conteo_numeros = df['Cifra'].value_counts()

# Calcular el total de apariciones de los números de interés
total_apariciones = sum(conteo_numeros[num] for num in numeros_interes)

# Calcular probabilidad para cada número de interés
probabilidades = {num: (conteo_numeros[num] / total_apariciones) * 100 for num in numeros_interes}

# Mostrar resultados
for num, prob in probabilidades.items():
    print(f"Probabilidad de que caiga el número {num}: {prob:.2f}%")
```

```python
import pandas as pd

# Datos proporcionados
data = {
    "Día": list(range(1, 39)),
    "Ultimas_2_cifras": [43, 4, 92, 27, 3, 33, 91, 61, 77, 63, 9, 6, 13, 33, 33, 21, 97, 74, 45, 95,
                         35, 72, 84, 47, 94, 48, 24, 64, 95, 28, 9, 92, 25, 92, 27, 59, 25, 46]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Identificar si el número termina en par o impar
df['Par_Impar'] = df['Ultimas_2_cifras'] % 2  # 0 para par, 1 para impar

# Determinar la decena (0-9, 10-19, etc.)
df['Decena'] = (df['Ultimas_2_cifras'] // 10) * 10

# Calcular probabilidad de cada decena
total_cifras = len(df)
probabilidades_decenas = df['Decena'].value_counts(normalize=True) * 100

# Calcular probabilidad de par/impar en cada decena
probabilidades_decenas_par_impar = df.groupby(['Decena', 'Par_Impar']).size().unstack(fill_value=0)
probabilidades_decenas_par_impar = (probabilidades_decenas_par_impar.div(probabilidades_decenas_par_impar.sum(axis=1), axis=0) * 100).round(2)

# Mostrar resultados
print("Probabilidad de cada decena (en %):")
print(probabilidades_decenas.sort_index())
print("\nProbabilidad de par/impar por decena (en %):")
print(probabilidades_decenas_par_impar.rename(columns={0: 'Par', 1: 'Impar'}))
```

```python
# Secuencia de números (la lista debe coincidir con los números de tu imagen)
numeros = [3, 4, 2, 7, 3, 3, 1, 1, 7, 3, 9, 6, 3, 3, 3, 1, 7, 4, 5, 5, 5, 2, 4, 7, 4, 8, 4, 4, 5, 8, 9, 2, 5]

# Función para contar pares e impares en la secuencia
def contar_par_impar(numeros):
    pares = sum(1 for num in numeros if num % 2 == 0)
    impares = len(numeros) - pares
    return pares, impares

# Contar pares e impares en la secuencia
pares, impares = contar_par_impar(numeros)

# Predicción simple basada en la tendencia
if pares > impares:
    prediccion = "Par"
else:
    prediccion = "Impar"

print(f"En la secuencia, hay {pares} números pares y {impares} números impares.")
print(f"Según la tendencia, se espera que el siguiente número sea: {prediccion}")
```

```python
import pandas as pd

# Datos proporcionados
data = {
    "Día": list(range(1, 39)),
    "Ultimas_2_cifras": [43, 4, 92, 27, 3, 33, 91, 61, 77, 63, 9, 6, 13, 33, 33, 21, 97, 74, 45, 95,
                         35, 72, 84, 47, 94, 48, 24, 64, 95, 28, 9, 92, 25, 92, 27, 59, 25, 46]
}

# Crear DataFrame
df = pd.DataFrame(data)

# 1. Calcular la Media Móvil Simple (SMA) con una ventana de 5 días
df['SMA_5'] = df['Ultimas_2_cifras'].rolling(window=5).mean()

# 2. Calcular la Media Móvil Exponencial (EMA) con una ventana de 5 días
df['EMA_5'] = df['Ultimas_2_cifras'].ewm(span=5, adjust=False).mean()

# 3. Calcular el Índice de Fuerza Relativa (RSI) con una ventana de 14 días
def calculate_rsi(series, period=14):
    delta = series.diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

df['RSI_14'] = calculate_rsi(df['Ultimas_2_cifras'])

# 4. Calcular la Desviación Estándar con una ventana de 5 días
df['STD_5'] = df['Ultimas_2_cifras'].rolling(window=5).std()

# Mostrar resultados
print(df[['Día', 'Ultimas_2_cifras', 'SMA_5', 'EMA_5', 'RSI_14', 'STD_5']])
```

```python
# Calcular el próximo valor usando el último valor de EMA
# Ajuste a la dirección en función del cruce entre SMA y EMA
ultimo_valor_ema = df['EMA_5'].iloc[-1]
ultimo_valor_sma = df['SMA_5'].iloc[-1]

# Estimar el siguiente valor basado en la dirección de la tendencia
if ultimo_valor_ema > ultimo_valor_sma:
    # Si EMA > SMA, hay tendencia alcista
    proximo_valor = ultimo_valor_ema + (ultimo_valor_ema - ultimo_valor_sma) * 0.5
else:
    # Si EMA < SMA, hay tendencia bajista
    proximo_valor = ultimo_valor_ema - (ultimo_valor_sma - ultimo_valor_ema) * 0.5

print(f"Proyección del próximo valor estimado: {proximo_valor:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados
data = {
    "Día": list(range(1, 39)),
    "Ultimas_2_cifras": [43, 4, 92, 27, 3, 33, 91, 61, 77, 63, 9, 6, 13, 33, 33, 21, 97, 74, 45, 95,
                         35, 72, 84, 47, 94, 48, 24, 64, 95, 28, 9, 92, 25, 92, 27, 59, 25, 46]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Calcular SMA (Media Móvil Simple) de 5 días
df['SMA_5'] = df['Ultimas_2_cifras'].rolling(window=5).mean()

# Calcular EMA (Media Móvil Exponencial) de 5 días
df['EMA_5'] = df['Ultimas_2_cifras'].ewm(span=5, adjust=False).mean()

# Calcular RSI (Índice de Fuerza Relativa) de 14 días
delta = df['Ultimas_2_cifras'].diff()
gain = np.where(delta > 0, delta, 0)
loss = np.where(delta < 0, -delta, 0)
average_gain = pd.Series(gain).rolling(window=14).mean()
average_loss = pd.Series(loss).rolling(window=14).mean()
rs = average_gain / average_loss
df['RSI_14'] = 100 - (100 / (1 + rs))

# Calcular Desviación Estándar (STD) de 5 días
df['STD_5'] = df['Ultimas_2_cifras'].rolling(window=5).std()

# Bandas de Bollinger
df['Upper_Band'] = df['SMA_5'] + (df['STD_5'] * 2)
df['Lower_Band'] = df['SMA_5'] - (df['STD_5'] * 2)

# MACD
df['EMA_12'] = df['Ultimas_2_cifras'].ewm(span=12, adjust=False).mean()
df['EMA_26'] = df['Ultimas_2_cifras'].ewm(span=26, adjust=False).mean()
df['MACD'] = df['EMA_12'] - df['EMA_26']
df['Signal_Line'] = df['MACD'].ewm(span=9, adjust=False).mean()

# Estocástico
L14 = df['Ultimas_2_cifras'].rolling(window=14).min()
H14 = df['Ultimas_2_cifras'].rolling(window=14).max()
df['Stochastic'] = 100 * (df['Ultimas_2_cifras'] - L14) / (H14 - L14)

# ATR (Average True Range)
df['High_Low'] = df['Ultimas_2_cifras'].rolling(window=2).max() - df['Ultimas_2_cifras'].rolling(window=2).min()
df['Close_Previous'] = df['Ultimas_2_cifras'].shift(1)
df['High_Close_Previous'] = (df['Ultimas_2_cifras'].rolling(window=2).max() - df['Close_Previous']).abs()
df['Low_Close_Previous'] = (df['Ultimas_2_cifras'].rolling(window=2).min() - df['Close_Previous']).abs()
df['True_Range'] = df[['High_Low', 'High_Close_Previous', 'Low_Close_Previous']].max(axis=1)
df['ATR'] = df['True_Range'].rolling(window=14).mean()

# Graficar
plt.figure(figsize=(14, 12))

# Gráfico de SMA, EMA y Bandas de Bollinger
plt.subplot(4, 1, 1)
plt.plot(df['Día'], df['Ultimas_2_cifras'], label='Ultimas_2_cifras', color='blue', marker='o')
plt.plot(df['Día'], df['SMA_5'], label='SMA 5', color='green', linestyle='--')
plt.plot(df['Día'], df['EMA_5'], label='EMA 5', color='red', linestyle='--')
plt.plot(df['Día'], df['Upper_Band'], label='Upper Band', color='orange', linestyle='--')
plt.plot(df['Día'], df['Lower_Band'], label='Lower Band', color='orange', linestyle='--')
plt.fill_between(df['Día'], df['Upper_Band'], df['Lower_Band'], color='lightgray', alpha=0.5)
plt.title('Serie de Ultimas_2_cifras con SMA, EMA y Bandas de Bollinger')
plt.xlabel('Día')
plt.ylabel('Valor')
plt.legend()

# Gráfico de MACD
plt.subplot(4, 1, 2)
plt.plot(df['Día'], df['MACD'], label='MACD', color='blue')
plt.plot(df['Día'], df['Signal_Line'], label='Signal Line', color='red')
plt.title('MACD')
plt.xlabel('Día')
plt.ylabel('MACD')
plt.axhline(0, color='black', linestyle='--')
plt.legend()

# Gráfico de Estocástico
plt.subplot(4, 1, 3)
plt.plot(df['Día'], df['Stochastic'], label='Estocástico', color='purple')
plt.axhline(80, color='red', linestyle='--', linewidth=0.5)
plt.axhline(20, color='green', linestyle='--', linewidth=0.5)
plt.title('Estocástico')
plt.xlabel('Día')
plt.ylabel('Estocástico')
plt.legend()

# Gráfico de ATR
plt.subplot(4, 1, 4)
plt.plot(df['Día'], df['ATR'], label='ATR', color='orange')
plt.title('Average True Range (ATR)')
plt.xlabel('Día')
plt.ylabel('ATR')
plt.legend()

plt.tight_layout()
plt.show()
```
